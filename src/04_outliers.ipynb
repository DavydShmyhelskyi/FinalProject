```r
# ================================
# 04 OUTLIERS ANALYSIS (FULL - FIXED)
# ================================

library(readr)
library(dplyr)
library(ggplot2)

setwd("C:/Users/Graf David/R/FinalProject")

df <- read_csv("dataset/train.csv", show_col_types = FALSE)
```

```r
important_vars <- c(
  "current price",
  "years",
  "km",
  "rating",
  "condition",
  "economy",
  "top speed",
  "hp",
  "torque"
)
```

```r
for (v in important_vars) {
  p <- ggplot(df, aes(y = .data[[v]])) +
    geom_boxplot(fill = "steelblue") +
    labs(title = paste("Boxplot of", v), y = v)
  
  file_name <- paste0("diagrams/box_", gsub(" ", "_", v), ".png")
  ggsave(file_name, plot = p, width = 6, height = 6)
  
  print(p)
}
```

```r
print_outliers_console <- function(x, name) {
  q1 <- quantile(x, 0.25, na.rm = TRUE)
  q3 <- quantile(x, 0.75, na.rm = TRUE)
  iqr <- q3 - q1
  
  lower <- q1 - 1.5 * iqr
  upper <- q3 + 1.5 * iqr
  
  outliers <- x[x < lower | x > upper]
  
  cat("\n=====================================\n")
  cat("OUTLIER REPORT FOR:", toupper(name), "\n")
  cat("=====================================\n")
  cat("Min:           ", min(x), "\n")
  cat("Q1:            ", q1, "\n")
  cat("Median:        ", median(x), "\n")
  cat("Q3:            ", q3, "\n")
  cat("Max:           ", max(x), "\n")
  cat("Lower bound:   ", lower, "\n")
  cat("Upper bound:   ", upper, "\n")
  cat("Outliers count:", length(outliers), "\n")
  
  if (length(outliers) == 0) {
    cat("Outliers values: NONE\n")
  } else {
    cat("First 10 outliers:\n")
    print(head(outliers, 10))
  }
}
```

```r
for (v in important_vars) {
  print_outliers_console(df[[v]], v)
}
```
