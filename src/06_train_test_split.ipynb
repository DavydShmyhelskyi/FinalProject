```r
# ================================
# 06 TRAIN / TEST SPLIT
# ================================

library(readr)
library(dplyr)

setwd("C:/Users/Graf David/R/FinalProject")

df <- read_csv("dataset/train.csv", show_col_types = FALSE)
```

```r
df_model <- df %>% select(-`v.id`)
```

```r
set.seed(42)
```

```r
sample_size <- floor(0.75 * nrow(df_model))

train_index <- sample(seq_len(nrow(df_model)), size = sample_size)

train_data <- df_model[train_index, ]
test_data  <- df_model[-train_index, ]
```

```r
dim(train_data)
dim(test_data)
```

```r
colMeans(train_data)
colMeans(test_data)
```

```r
# ================================
# FULL SUMMARY — TRAIN / TEST SPLIT
# ================================

cat("\n==============================\n")
cat("FULL DATA SPLIT SUMMARY\n")
cat("==============================\n")

cat("\n--- DATASET SHAPES ---\n")
cat("Original dataset:", dim(df), "\n")
cat("Model dataset (without v.id):", dim(df_model), "\n")
cat("Train dataset:", dim(train_data), "\n")
cat("Test dataset:", dim(test_data), "\n")

cat("\n--- TRAIN MEANS ---\n")
print(round(colMeans(train_data), 2))

cat("\n--- TEST MEANS ---\n")
print(round(colMeans(test_data), 2))

cat("\n--- TRAIN SD ---\n")
print(round(sapply(train_data, sd), 2))

cat("\n--- TEST SD ---\n")
print(round(sapply(test_data, sd), 2))

cat("\n--- TARGET VARIABLE CHECK (current price) ---\n")
cat("Train price mean:", mean(train_data$`current price`), "\n")
cat("Test  price mean:", mean(test_data$`current price`), "\n")

cat("\n==============================\n")
cat("END OF STEP 06\n")
cat("==============================\n")
```
