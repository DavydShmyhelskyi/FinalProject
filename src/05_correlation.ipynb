```r
# ================================
# 05 CORRELATION ANALYSIS
# ================================

library(readr)
library(dplyr)
library(corrplot)
library(ggplot2)

setwd("C:/Users/Graf David/R/FinalProject")

df <- read_csv("dataset/train.csv", show_col_types = FALSE)
```

```r
corr_vars <- df %>%
  select(
    `current price`,
    `on road old`,
    `on road now`,
    years,
    km,
    rating,
    condition,
    economy,
    `top speed`,
    hp,
    torque
  )
```

```r
corr_matrix <- cor(corr_vars)
corr_matrix
```

```r
png("diagrams/correlation_heatmap.png", width = 900, height = 800)

corrplot(
  corr_matrix,
  method = "color",
  type = "upper",
  tl.col = "black",
  tl.cex = 0.8,
  number.cex = 0.7
)

dev.off()
```

```r
price_col_index <- which(colnames(corr_matrix) == "current price")
cor_with_price <- sort(corr_matrix[, price_col_index], decreasing = TRUE)
print(cor_with_price)
```
