```r
# ================================
# 02 DESCRIPTIVE STATISTICS
# ================================

library(readr)
library(dplyr)

setwd("C:/Users/Graf David/R/FinalProject")

df <- read_csv("dataset/train.csv")
```

```r
# Загальна описова статистика для всіх змінних
summary(df)
```

```r
# Середнє значення
sapply(df, mean)

# Стандартне відхилення
sapply(df, sd)
```

```r
df %>%
  summarise(
    min_price = min(`current price`),
    q1_price  = quantile(`current price`, 0.25),
    median_price = median(`current price`),
    mean_price   = mean(`current price`),
    q3_price  = quantile(`current price`, 0.75),
    max_price = max(`current price`),
    sd_price  = sd(`current price`)
  )
```

```r
df %>%
  summarise(
    min_years = min(years),
    q1_years  = quantile(years, 0.25),
    median_years = median(years),
    mean_years   = mean(years),
    q3_years  = quantile(years, 0.75),
    max_years = max(years),
    sd_years  = sd(years)
  )
```

```r
df %>%
  summarise(
    min_km = min(km),
    q1_km  = quantile(km, 0.25),
    median_km = median(km),
    mean_km   = mean(km),
    q3_km  = quantile(km, 0.75),
    max_km = max(km),
    sd_km  = sd(km)
  )
```

```r
# ================================
# FULL DATASET INFORMATION
# ================================

cat("===== DATASET DIMENSIONS =====\n")
print(dim(df))

cat("\n===== COLUMN NAMES =====\n")
print(colnames(df))

cat("\n===== DATA STRUCTURE (str) =====\n")
str(df)

cat("\n===== GLIMPSE (dplyr) =====\n")
glimpse(df)

cat("\n===== MISSING VALUES PER COLUMN =====\n")
print(colSums(is.na(df)))

cat("\n===== DATA TYPES =====\n")
print(sapply(df, class))
```
