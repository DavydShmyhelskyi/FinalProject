```r
# ================================
# 03 DISTRIBUTIONS
# ================================

library(readr)
library(dplyr)
library(ggplot2)

setwd("C:/Users/Graf David/R/FinalProject")

df <- read_csv("dataset/train.csv", show_col_types = FALSE)
```

```r
p1 <- ggplot(df, aes(x = `current price`)) +
  geom_histogram(bins = 30, fill = "steelblue", color = "black") +
  labs(
    title = "Distribution of Current Price",
    x = "Current Price",
    y = "Count"
  )

ggsave("diagrams/hist_current_price.png", plot = p1, width = 7, height = 5)

p1
```
